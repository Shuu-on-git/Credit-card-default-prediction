# Create the TranspositionCipher class

```python
class TranspositionCipher(object): 
        
    def __init__(self, key):
        self.key = key
        
    def encrypt_message(self, message):
        text = list(message)
        encry = ""
        wid = self.key
        lis = []
        null = '-------'

        for i in range(0,len(text),wid):
            '''this loop is for breaking the string in to the given key: 
            - converted list of strings to list of characters 
            - used split to split the text string
            '''
            x = text[i:i + wid]
        
            
            
            lis.append(list(x))


        for i in range(wid - (len(lis[-1]))) :
            '''adds null values at the end of unequal list'''
            lis[-1].append(null)

        for i in range(0,wid):
            '''This is the encryption loop'''
            for j in range(0,len(lis)):
                if lis[j][i] != null:
                    encry += (lis[j][i])
                    
        print(encry)
        self.encry = encry
    
    def decrypt_message(self, message):
        text = list(message)
        decry = ""
        lis = []
        null = '-------'
        wid = self.key


        if (len(text) % wid!= 0 ): 
            '''dividing this into 2 parts : 

            - full list of characters 
            - with empty character list 
            '''    


            nwid = ((len(text)//wid) + 1)

            dimx = nwid * wid # dimension of the table created 


            # length of the lower half
            emptycell = dimx - len(text) 


            # list of starting loc of upper list  
            uploc = list(range(0,dimx,nwid))

            uploc = uploc[:-emptycell]


            # list of starting loc of lower list 
            firstoflow = uploc[-1] + nwid
            loflower = nwid - 1 # as these rows have empty cells
            lowloc = list(range(firstoflow,len(text),loflower))   


            # list of starting loc of both upper and lower parts
            startloc = uploc + lowloc

            for i in startloc:
                '''This loop is for the separating list full of characters and list with empty characters'''
                if i in uploc :
                    x = list(text[i:i+nwid])
                    lis.append(x)
                else :
                    x = list(text[i: i+nwid-1])
                    x.append(null)
                    lis.append(x)

        else : 
            nwid = (len(text)//wid)

            dimx = nwid * wid 
            loc = list(range(0,dimx,nwid))


            for i in loc:
                x = list(text[i: i+nwid])
                lis.append(x)





        for i in range(0,nwid):
            '''This is the decryption loop'''
            for j in range(0,len(lis)):
                if lis[j][i] != null:
                    decry += (lis[j][i])


        
            
        self.decry = decry
        
        print(decry)
```
